```python
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
```

```python
#Remove Readme
os.remove('../Datasets/COVID-19_Radiography_Dataset/README.md.txt')
```

```python
#Utilizing list comprehension to create a list that will get the relative file paths for each folder
features = ['COVID','Lung_Opacity','Normal','Viral Pneumonia']
image_file_path = [f'../Datasets/COVID-19_Radiography_Dataset/{feature}/images/' for feature in features]
```

```python
#Collected image file paths for all images across all folders
file_list= []
file_name = []
for feature in image_file_path:
    images = os.listdir(feature)
    for image in images:
        file_name.append(image)
        file_list.append(feature + image)
```

```python
#Used OpenCvs imread to turn all images into an array
image_list = []
for png in file_list:
    image_list.append(cv2.imread(png,0))
```

```python
#turn list into array
image_list = np.array(image_list)
```
